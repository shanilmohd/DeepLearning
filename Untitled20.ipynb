```python
from fastapi import FastAPI, Response
from starlette.middleware import Middleware
from starlette.middleware.cors import CORSMiddleware
import torch
from torch import autocast
from diffusers import StableDiffusionPipeline
from io import BytesIO
import base64
```

```python


middleware = [
    Middleware(
    CORSMiddleware,
     allow_origins=['http://192.168.43.67:3000'],
    )
  ]
app = FastAPI(middleware=middleware)
device="cuda"
model_id="stabilityai/stable-diffusion-2-1"
pipe=StableDiffusionPipeline.from_pretrained(model_id, revision="fp16", torch_dtype=torch.float16, use_auth_token="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")
pipe.to(device)
@app.get("/")
def generate(prompt: str):
  with autocast(device):
    image=pipe(prompt,num_inference_steps=75).images[0]
  image.save("testImage.png")
  buffer=BytesIO()
  image.save(buffer, format="PNG")
  imgstr=base64.b64encode(buffer.getvalue())
  return Response(content=imgstr, media_type="image/png")
```
